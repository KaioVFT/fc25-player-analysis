```python
# Calculate the 10 players with the highest overall average or 'OVR'

df_bestOVR = df.sort_values(by='OVR', ascending=False)
df_bestOVR.head(10)
```

```python
# Calculate the 10 players with the lowest overall average or 'OVR'

df_worstOVR = df.sort_values(by='OVR', ascending=False)
df_worstOVR.tail(10)
```

```python
# Calculate the 20 players with the highest estimated potential or 'Potential' under 20 years old

# Filtrar jogadores com idade menor que 20 anos
df_under_20 = df[df["Age"] <= 20]

# Criar a coluna de Potencial Estimado
df_under_20["Potential_Estimate"] = df_under_20["OVR"] + (25 - df_under_20["Age"]) * 0.5

# Garante que o potencial m√≠nimo seja o OVR atual (evita valores menores)
df_under_20["Potential_Estimate"] = df_under_20[["OVR", "Potential_Estimate"]].max(axis=1)

# Ordenar os jogadores com maior potencial estimado
top_potential = df_under_20.sort_values(by="Potential_Estimate", ascending=False)

# Exibir os 20 jogadores com maior potencial estimado
print(top_potential[["Name", "Age", "OVR", "Potential_Estimate"]].head(20))
```

```python
# Calculate the 20 players with the highest estimated potential regardless of age

# Criar a coluna de Potencial Estimado
df["Potential_Estimate"] = df["OVR"] + (25 - df["Age"]) * 0.5

# Garante que o potencial m√≠nimo seja o OVR atual (evita valores menores)
df["Potential_Estimate"] = df[["OVR", "Potential_Estimate"]].max(axis=1)

# Ordenar os jogadores com maior potencial estimado
top_potential = df.sort_values(by="Potential_Estimate", ascending=False)

# Exibir os 20 jogadores com maior potencial estimado
print(top_potential[["Name", "Age", "OVR", "Potential_Estimate"]].head(20))
```
